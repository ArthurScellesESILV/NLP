```python
import os

def read_reviews(file_path):
    reviews = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            reviews.append(line.strip())
    return reviews

# Chemins vers les dossiers contenant les fichiers .txt
neg_folder = '/content/neg'
pos_folder = '/content/pos'

# Spécifiez un fichier dans chaque dossier
neg_files = [os.path.join(neg_folder, file) for file in os.listdir(neg_folder) if file.endswith('.txt')]
pos_files = [os.path.join(pos_folder, file) for file in os.listdir(pos_folder) if file.endswith('.txt')]

# Charger les avis négatifs et positifs
neg_reviews = [read_reviews(file) for file in neg_files]
pos_reviews = [read_reviews(file) for file in pos_files]

# Afficher quelques avis pour vérification
print("Exemple d'avis négatif :", neg_reviews[0][0])
print("Exemple d'avis positif :", pos_reviews[0][0])
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag
from nltk.corpus import wordnet as wn

# Assurez-vous que les ressources nécessaires sont téléchargées
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def pos_tag_reviews(reviews):
    pos_tagged_reviews = []
    for review in reviews:
        # Convertir l'élément en chaîne de caractères
        review_str = str(review)

        sentences = sent_tokenize(review_str)
        pos_tags_per_review = []
        for sentence in sentences:
            words = word_tokenize(sentence)
            pos_tags_per_sentence = pos_tag(words)
            pos_tags_per_review.extend(pos_tags_per_sentence)
        pos_tagged_reviews.append(pos_tags_per_review)
    return pos_tagged_reviews



# Appliquer le pos-tagging aux avis négatifs et positifs
neg_pos_tagged = pos_tag_reviews(neg_reviews)
pos_pos_tagged = pos_tag_reviews(pos_reviews)

# Afficher quelques tags pour vérification
print("Exemple de pos-tags dans les avis négatifs :", neg_pos_tagged[:10])
print("Exemple de pos-tags dans les avis positifs :", pos_pos_tagged[:10])
```

```python
def extract_adverbs(pos_tagged_reviews):
    adverbs = []
    for pos_tags in pos_tagged_reviews:
        for word, pos in pos_tags:
            if pos == 'RB':  # RB est le tag pour les adverbes
                adverbs.append(word)
    return adverbs


neg_adverbs = extract_adverbs(neg_pos_tagged)
pos_adverbs = extract_adverbs(pos_pos_tagged)


print("Adverbes dans les avis négatifs :", neg_adverbs[:10])
print("Adverbes dans les avis positifs :", pos_adverbs[:10])
```

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag
from nltk.corpus import sentiwordnet as swn

# Assurez-vous que la ressource WordNet est téléchargée
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('sentiwordnet')
```

```python
from nltk.corpus import wordnet as wn

def get_sentiment_score(word, pos):
    # Obtient tous les synsets correspondant au mot et à la partie du discours (pos)
    synsets = list(wn.synsets(word, pos))

    if not synsets:
        return None

    # Prend le premier synset (c'est une simplification, vous pouvez choisir d'autres stratégies)
    synset = synsets[0]

    # Obtient les scores de sentiment de SentiWordNet
    sentiment = swn.senti_synset(synset.name())

    return sentiment

word = 'pretty'
pos = 'r'  # 'r' pour adverbe
sentiment_score = get_sentiment_score(word, pos)

# Imprime le score de sentiment
print(f"Score de sentiment pour '{word}': Positif={sentiment_score.pos_score()}, Négatif={sentiment_score.neg_score()}")
```

```python
def classify_review(adverbs, threshold=0.0):
    total_pos_score = 0
    total_neg_score = 0

    for adverb in adverbs:
        sentiment_score = get_sentiment_score(adverb, 'r')
        if sentiment_score:
            total_pos_score += sentiment_score.pos_score()
            total_neg_score += sentiment_score.neg_score()

    overall_score = total_pos_score - total_neg_score

    # Classification basée sur un seuil
    if overall_score > threshold:
        return "Positif"
    elif overall_score < -threshold:
        return "Négatif"
    else:
        return "Neutre"

# Classifie les avis négatifs
neg_classification = classify_review(neg_adverbs)

# Classifie les avis positifs
pos_classification = classify_review(pos_adverbs)

# Imprime les résultats
print("Classification des avis négatifs :", neg_classification)
print("Classification des avis positifs :", pos_classification)
```
